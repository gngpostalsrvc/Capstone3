```python
import json
import pandas as pd
from transformers import AutoTokenizerFast
from tokenizers import ByteLevelBPETokenizer, Tokenizer, models, pre_tokenizers, trainers
```

```python
df = pd.read_csv('HebrewBiblebyVerse.csv', index_col='Verse')

df.tail()
```

```python
verses = df['Text'].tolist()

verses
```

```python
def get_training_corpus():
    dataset = verses
    for start_idx in range(0, len(dataset), 1000):
        samples = dataset[start_idx : start_idx + 1000]
        yield samples
        
training_corpus = get_training_corpus()
```

```python
old_tokenizer = AutoTokenizerFast.from_pretrained('gpt2')
tokenizer = old_tokenizer.train_new_from_iterator(training_corpus, 52000)
```

```python
tokenizer.save_pretrained(".", "BERiT")
```

```python
with open('tokenizer_config.json', 'r') as vocab_json:
    tokens = json.load(vocab_json)
```

```python
encoding = tokenizer.encode(verses[0])
```

```python
tokenizer.decode(encoding)
```

```python
tokenizer = Tokenizer(models.BPE())
tokenizer.pre_tokenizer = pre_tokenizers.ByteLevel(add_prefix_space=False)
trainer = trainers.BpeTrainer(vocab_size=25000, special_tokens=["<|endoftext|>"])

tokenizer.train_from_iterator(get_training_corpus(), trainer=trainer)
```

```python
tokenizer.save_pretrained(".", "BERiT")
```
